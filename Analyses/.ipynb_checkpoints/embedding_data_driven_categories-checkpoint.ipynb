```python
# --------------------------------------------------
# Imports
# --------------------------------------------------
import pandas as pd
import ast
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import AgglomerativeClustering
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from sklearn.cluster import AgglomerativeClustering
from huggingface_hub import InferenceClient
from collections import Counter, defaultdict
import re

# --------------------------------------------------
# 1. Load Extracted Functions and Select Most Frequents
# --------------------------------------------------
print("Loading extracted LLM functions...")
df = pd.read_csv('../Output/llm_generated_deliberation_functions.csv')

function_set = set()
for i, row in df.iterrows():
    raw = row.get('functions_extracted')
    if pd.isna(raw) or raw.strip() == "NA":
        continue
    try:
        parsed = ast.literal_eval(raw)
        for func in parsed:
            operation = func.get("operation", "").strip().lower()
            if operation and operation != "na":
                function_set.add(operation)
    except Exception as e:
        print(f"Skipping row {i} due to parsing error: {e}")

unique_functions = sorted(function_set)
pd.Series(unique_functions, name="function").to_csv('../Output/unique_cognitive_functions.csv', index=False)
print(f"Extracted {len(unique_functions)} unique functions.")

# Count occurrences across all rows
# Count all function occurrences
function_counter = Counter()

for i, row in df.iterrows():
    raw = row.get('functions_extracted')
    if pd.isna(raw) or raw.strip() == "NA":
        continue
    try:
        parsed = ast.literal_eval(raw)
        for func in parsed:
            operation = func.get("operation", "").strip().lower()
            if operation and operation != "na":
                function_counter[operation] += 1
    except Exception:
        continue  # already logged above

# Create full count DataFrame
function_counts_df = pd.DataFrame.from_dict(function_counter, orient='index', columns=['count'])
function_counts_df.index.name = 'function'
function_counts_df = function_counts_df.sort_values(by='count', ascending=False)

# Filter: count ≥ 10
frequent_functions_df = function_counts_df[function_counts_df['count'] >= 10]

# Save
frequent_functions_df.to_csv('../Output/frequent_functions_min10.csv')

# Display
print("\nFunctions with at least 10 occurrences:")
print(frequent_functions_df)
```

```python
# --------------------------------------------------
# 2. Embedding (for frequent functions only)
# --------------------------------------------------
print("Embedding frequent cognitive functions (count ≥ 10)...")

frequent_functions = list(frequent_functions_df.index)
function_texts = ["clustering: " + f for f in frequent_functions]

model = SentenceTransformer("nomic-ai/nomic-embed-text-v1.5", trust_remote_code=True)
embeddings = model.encode(function_texts, convert_to_tensor=True)
embeddings = F.normalize(embeddings, p=2, dim=1).cpu().numpy()

print("Embeddings complete.")

# --------------------------------------------------
# 3. Cosine Similarity Matrix
# --------------------------------------------------
print("Computing cosine similarity matrix for frequent functions...")
sim_matrix = cosine_similarity(embeddings)
sim_df = pd.DataFrame(sim_matrix, index=frequent_functions, columns=frequent_functions)
sim_df.to_csv("../Output/cosine_similarity_matrix_frequent.csv")

print("Cosine similarity matrix saved.")
```

```python
import time

# --------------------------------------------------
# LLaMA 3.3 setup
# --------------------------------------------------
client = InferenceClient(
    provider="together",
    api_key="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",
)
LLAMA_MODEL   = "meta-llama/Llama-3.3-70B-Instruct"
temperature   = 0
top_p         = 0
max_tokens    = 300
n             = 1
MAX_RETRIES   = 3          # maximum attempts per call
RETRY_DELAY   = 60         # seconds to wait on failure


def get_simplified_labels(terms: list[str]) -> str:
    """Send a two‑term list to LLaMA and return clean text. Retries up to MAX_RETRIES."""
    
    prompt = f"""
You are a cognitive scientist specializing in the taxonomy of reasoning processes.

You are given a set of two cognitive function labels.

Your task:
- Examine the two terms below.
- If the terms refer to the same underlying mental operation, merge them under a **function label** (1–2 words).
- If the terms refer to different mental operations, list them separately, each under its own function label.
- Do not use overly general labels such as "reasoning", "thinking", "deliberation", or "decision‑making".
- Do not include any commentary or explanation.
- Output only in the following format:

FunctionLabel: [list of merged terms]

Only include meaningful groupings. If the terms do not belong together, return two separate lines using the same format — one for each term.

Terms:
{chr(10).join(['- ' + t for t in terms])}
""".strip()

    attempt = 0
    while attempt < MAX_RETRIES:
        try:
            response = client.chat.completions.create(
                model     = LLAMA_MODEL,
                messages  = [{"role": "user", "content": prompt}],
                temperature = temperature,
                top_p       = top_p,
                max_tokens  = max_tokens,
                n           = n
            )
            return response.choices[0].message.content.strip()

        except Exception as e:
            attempt += 1
            print(f"LLaMA call failed (attempt {attempt}/{MAX_RETRIES}): {e}")
            if attempt < MAX_RETRIES:
                print(f"Waiting {RETRY_DELAY} seconds before retrying…")
                time.sleep(RETRY_DELAY)
            else:
                print("Maximum retries reached; skipping this pair.")
                return ""
```

```python
# --------------------------------------------------
# Compute Cosine‑Similarity Threshold (Q3)
# --------------------------------------------------
print("Preparing similarity‑based LLM simplification…")

upper_idx = np.triu_indices_from(sim_matrix, k=1)
similarity_vals = sim_matrix[upper_idx]
q3_threshold = np.quantile(similarity_vals, 0.75)

print(f"Q3 cosine‑similarity threshold: {q3_threshold:.3f}")

# --------------------------------------------------
# LLM‑Based Merging of Highly Similar Pairs
# --------------------------------------------------
merged_results = []
processed_pairs = set()
merge_map = {}                         # term -> abstract label

def parse_llm_output(text: str) -> dict:
    """Return {original_term: abstract_label} from a clean LLaMA response."""
    mapping = {}
    for line in text.strip().splitlines():
        m = re.match(r'^([A-Za-z0-9_ -]+):\s*\[(.*?)\]$', line.strip())
        if m:
            abstract = m.group(1).strip().lower()
            terms = [t.strip().lower() for t in m.group(2).split(",")]
            for t in terms:
                mapping[t] = abstract
    return mapping

for i, f1 in enumerate(frequent_functions):
    for j, f2 in enumerate(frequent_functions):
        if i >= j:
            continue
        sim = sim_matrix[i, j]
        if sim > q3_threshold and (f1, f2) not in processed_pairs:
            processed_pairs.update({(f1, f2), (f2, f1)})
            print(f"\n▶ Submitting pair (cos sim = {sim:.2f}): [{f1}]  &  [{f2}]")

            llm_resp = get_simplified_labels([f1, f2])
            print("LLM response:")
            print(llm_resp)

            merged_results.append(
                {"pair": f"{f1} | {f2}", "similarity": sim, "llm_output": llm_resp}
            )

            merge_map.update(parse_llm_output(llm_resp))

# --------------------------------------------------
# Apply Merges to Function List
# --------------------------------------------------
print("\nApplying LLM‑based merges…")

final_functions = [merge_map.get(func, func) for func in frequent_functions]
final_functions_unique = sorted(set(final_functions))

print(f"\nFinal unique functions after merge: {len(final_functions_unique)}")
print(final_functions_unique)

# --------------------------------------------------
# Save Logs
# --------------------------------------------------
pd.DataFrame(merged_results).to_csv(
    "../Output/llm_pairwise_merge_log.csv", index=False
)
pd.Series(final_functions_unique, name="function").to_csv(
    "../Output/llm_simplified_functions_final.csv", index=False
)

print("\nMerge log and simplified list saved.")
```

```python
# --------------------------------------------------
# 1–3. Extract valid abstract labels and deduplicate
# --------------------------------------------------
print("Extracting abstract labels from LLM output...")

valid_lines = []
invalid_lines = []

for response in simplified_df["llm_output"]:
    if pd.isna(response):
        continue
    for line in response.strip().splitlines():
        if re.match(r'^[A-Za-z0-9_ -]+:\s*\[.*?\]$', line.strip()):
            valid_lines.append(line.strip())
        elif line.strip():
            invalid_lines.append(line.strip())

if invalid_lines:
    print(f"{len(invalid_lines)} invalid lines found:")
    for line in invalid_lines:
        print(f"- {line}")

# Extract labels
abstract_labels = [re.match(r'^([A-Za-z0-9_ -]+):\s*\[.*?\]$', line).group(1).strip().lower()
                   for line in valid_lines]

# Deduplicate
abstract_labels_unique = sorted(set(abstract_labels))
print(f"\nRetained {len(abstract_labels_unique)} unique abstract labels.")
```

```python
# --------------------------------------------------
# Second-Round Merge: All Abstract Labels at Once
# --------------------------------------------------
print("Submitting all abstract labels to LLaMA for a second-round conceptual merge...")

abstract_labels_list = sorted(abstract_labels_unique)

second_round_prompt = f"""
You are a cognitive scientist specializing in the taxonomy of reasoning processes.

You are given a list of cognitive function labels.

Your task:
- Examine the list of terms below.
- If any refer to the same or similar underlying mental operation, merge them under a **higher-level abstract function label** (1–2 words).
- Do not use labels that are too general such as "Reasoning"
- Do not provide any explanation, commentary, or additional text.
- Output only in the format below:

Correct format:
Abstract_Label: [merged_term_1, merged_term_2, ...]

Terms:
{chr(10).join(['- ' + label for label in abstract_labels_list])}
"""

try:
    response = client.chat.completions.create(
        model=LLAMA_MODEL,
        messages=[{"role": "user", "content": second_round_prompt}],
        temperature=temperature,
        top_p=top_p,
        max_tokens=max_tokens,
        n=n
    )
    second_round_output = response.choices[0].message.content.strip()
    print("LLaMA 3.3 second-round merge response:\n")
    print(second_round_output)
except Exception as e:
    print(f"Second-round LLaMA merge failed: {e}")
    second_round_output = ""
```

```python
# --------------------------------------------------
# 1. Extract Valid Merged Labels from LLaMA Output
# --------------------------------------------------
print("Extracting valid lines from second-round LLaMA output...")

valid_lines = []
invalid_lines = []
pattern = r'^([A-Za-z0-9_ -]+):\s*\[.*?\]$'

for line in second_round_output.strip().splitlines():
    if re.match(pattern, line.strip()):
        valid_lines.append(line.strip())
    elif line.strip():
        invalid_lines.append(line.strip())

if invalid_lines:
    print(f"\n{len(invalid_lines)} invalid lines removed:")
    for line in invalid_lines:
        print(f"- {line}")

# --------------------------------------------------
# 2. Deduplicate Abstract Labels
# --------------------------------------------------
abstract_labels_final = sorted(set(
    re.match(pattern, line).group(1).strip().lower()
    for line in valid_lines
))

print(f"\nFinal deduplicated abstract labels ({len(abstract_labels_final)}):")
print(abstract_labels_final)

label_df = pd.DataFrame({"abstract_label": abstract_labels_final})

# --------------------------------------------------
# 3. Embed Final Labels
# --------------------------------------------------
print("Embedding final abstract labels...")
texts = ["clustering: " + label for label in label_df["abstract_label"]]
embeddings = model.encode(texts, convert_to_tensor=True)
embeddings = F.normalize(embeddings, p=2, dim=1).cpu().numpy()

# --------------------------------------------------
# 4. UMAP Projection
# --------------------------------------------------
print("Reducing embeddings to 2D with UMAP...")
umap = UMAP(n_components=2, n_neighbors=5, metric="cosine", random_state=42)
coords = umap.fit_transform(embeddings)

label_df["x"] = coords[:, 0]
label_df["y"] = coords[:, 1]

# --------------------------------------------------
# 5. Final Plot
# --------------------------------------------------
print("Plotting...")
plt.figure(figsize=(10, 8))
plt.scatter(label_df["x"], label_df["y"], s=0, alpha=0)

for _, row in label_df.iterrows():
    plt.text(row["x"], row["y"], row["abstract_label"], fontsize=9, alpha=0.9)

plt.title("2D UMAP Projection of Final Abstract Cognitive Categories")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.grid(True, linestyle="--", alpha=0.3)
sns.despine()
plt.tight_layout()
plt.savefig("../Output/final_llm_abstract_labels_umap.png", dpi=300)
plt.show()

print("Done.")
```
